#### 27. ‘노선명’을 행으로, 사용일자1의 요일(weekday)을 열로 하여 승차총승객수 평균을 가진 피벗 테이블 생성하기

`pivot_table`

```python
subway_df['요일'] = subway_df['사용일자1'].dt.day_name()
pivot_table = subway_df.pivot_table(values='승차총승객수',
                           index='노선명',
                           columns='요일',
                           aggfunc='mean',
                           fill_value=0)
pivot_table
```

#### 29. 28번에서 만든 피벗 테이블을 stack하여 긴 형태(Long form)로 변환하기

`stack`, `reset_index`

```python
stacked_data = pivot_table.stack()
stacked_df = stacked_data.reset_index()
stacked_df.columns = ['노선명', '요일', '평균승차인원']
stacked_df.head()
```

#### 29. 승차총승객수와 하차총승객수 컬럼을 melt를 이용해 ‘구분’(승차/하차)과 ‘인원’ 컬럼으로 변환하기

`melt`

```python
melted_data = pd.melt(subway_df,
                     id_vars=['노선명', '역명', '사용일자1'],
                     value_vars=['승차총승객수', '하차총승객수'],
                     var_name='구분',
                     value_name='인원')
melted_data['구분'] = melted_data['구분'].map({
    '승차총승객수': '승차',
    '하차총승객수': '하차'
})

melted_data.head()
```

#### 30. 27번에서 생성한 멀티 인덱스의 레벨 순서를 바꾸고, 특정 레벨을 제거한 후 xs 함수로 데이터를 추출하세요.

`swaplevel`, `droplevel`, `xs`


```python
multi_index_df = subway_df.set_index(['노선명', '역명'])
swapped_df = multi_index_df.swaplevel(0, 1).sort_index()
dropped_df = swapped_df.droplevel('노선명')
gangnam_data = swapped_df.xs('강남', level='역명')
print("레벨 순서 변경:")
display(swapped_df.head())
print("강남역 데이터:") 
display(gangnam_data.head())
```


#### 31. 평균 '승차총승객수'가 80,000명 이상이면서 동시에 '하차총승객수'의 표준편차가 30,000 이하인 노선 그룹만 필터링하세요.

`groupby`, `filter`, `lambda`

```python
filtered_lines = subway_df.groupby('노선명').filter(
    lambda x: (x['승차총승객수'].mean() >= 80000) & 
              (x['하차총승객수'].std() <= 30000)
)

print(f"필터링된 노선:) {filtered_lines['노선명'].unique()}")
print(f"필터링된 데이터 수: {len(filtered_lines)}")

filtered_lines.head()
```

#### 32. ‘승차총승객수’를 구간별로 나누어 ‘승차구간’이라는 새로운 컬럼을 추가하세요.
- 구간은 0-50,000, 50,001-100,000, 100,001-150,000, 150,001 이상으로 설정합니다.
- `cut`

```python
bins = [0, 25000, 50000, 750000, float('inf')]
labels = ['0-25K', '25K-50K', '50-75K', '75K+']
subway_df['승차구간'] = pd.cut(subway_df['승차총승객수'], bins=bins, labels=labels, right=False)
subway_df[['승차총승객수', '승차구간']]
```
