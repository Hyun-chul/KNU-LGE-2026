### 2. 그룹 분석 (GroupBy)
#### 2-1. 기본 그룹 분석

```python
# SAC 패턴 단계별 이해
print("=== Split-Apply-Combine 패턴 ===")

# 1. Split: 데이터 분할 과정 시각화
grouped = df.groupby('smoker')
print("1. Split - 그룹 분할:")
print(f"전체 그룹 수: {grouped.ngroups}")
print(f"그룹 키: {list(grouped.groups.keys())}")
print(f"각 그룹 크기: {grouped.size()}")

# 2. Apply: 각 그룹에 함수 적용
print("\n2. Apply - 함수 적용:")
for name, group in grouped:
    print(f"{name} 그룹 평균 팁: {group['tip'].mean():.2f}")

# 3. Combine: 결과 결합
print("\n3. Combine - 결과 결합:")
result = grouped['tip'].mean()
print(result)
```

#### 2-2. 다양한 집계 함수 활용

```python
# 기본 집계 함수들
print("기본 집계 함수들:")
print("합계:", df.groupby('smoker')['tip'].sum())
print("평균:", df.groupby('smoker')['tip'].mean())
print("중앙값:", df.groupby('smoker')['tip'].median())
print("최소값:", df.groupby('smoker')['tip'].min())
print("최대값:", df.groupby('smoker')['tip'].max())
print("개수:", df.groupby('smoker')['tip'].count())
print("표준편차:", df.groupby('smoker')['tip'].std())

# 다중 그룹화
print("\n다중 그룹화:")
gb = df.groupby(['smoker', 'time'])
out = gb[['total_bill', 'tip']].agg('mean')
print(out)
```

#### 2-3. 고급 집계 함수

```python
# 다양한 집계 함수 한 번에 적용
grouped = df.groupby('day').agg({
    'total_bill': ['sum', 'mean', 'min', 'max'],
    'tip': ['mean', 'std', 'count']
})
print("다양한 집계 함수 적용:")
display(grouped)
```

```python
# Named Aggregation
result = df.groupby('smoker').agg(
    total_bill_sum=('total_bill', 'sum'),
    tip_avg=('tip', 'mean'),
    tip_std=('tip', 'std')
)
print("\nNamed Aggregation:")
display(result)
```

```python
# reset_index로 인덱스 정리
grouped = df.groupby(['day', 'sex'])['tip'].mean()
grouped_reset = grouped.reset_index()
print("\nreset_index 사용:")
display(grouped_reset)
```

#### 2-4. 고급 그룹화 함수 - transform() 함수

```python
# transform() 함수 활용
print("transform() 함수 활용:")

# 그룹별 평균 대비 비율
df['tip_ratio_to_group_mean'] = df.groupby('day')['tip'].transform(lambda x: x / x.mean())

# 그룹별 표준화 (Z-score)
df['tip_zscore'] = df.groupby('day')['tip'].transform(lambda x: (x - x.mean()) / x.std())

# 그룹별 순위
df['tip_rank_in_day'] = df.groupby('day')['tip'].transform('rank')

# 그룹별 누적합
df['tip_cumsum_in_day'] = df.groupby('day')['tip'].transform('cumsum')

display(df[['day', 'tip', 'tip_ratio_to_group_mean', 'tip_zscore', 'tip_rank_in_day']].head(10))
```

```python
# 결측값 처리에 transform 활용
print("\n결측값 처리:")
df_with_na = df.copy()
df_with_na.loc[0:2, 'tip'] = None
df_with_na['tip_filled'] = df_with_na.groupby('day')['tip'].transform(
    lambda x: x.fillna(x.mean())
)
display(df_with_na[['day', 'tip', 'tip_filled']].head())
```

#### 2-4. 고급 그룹화 함수 - filter() 함수

```python
# filter() 함수 활용
print("filter() 함수 활용:")

# 그룹 크기가 30개 이상인 요일만 선택
large_day_groups = df.groupby('day').filter(lambda x: len(x) >= 30)
print(f"대규모 그룹 요일 데이터 수: {len(large_day_groups)}")
print(f"포함된 요일: {large_day_groups['day'].unique()}")
```

```python
# 평균 계산서가 20달러 이상인 요일만 선택
high_bill_days = df.groupby('day').filter(lambda x: x['total_bill'].mean() > 20)
print(f"고액 결제 요일 데이터 수: {len(high_bill_days)}")
print(f"포함된 요일: {high_bill_days['day'].unique()}")
```

```python
# 복합 조건 필터링
complex_filter = df.groupby('day').filter(
    lambda x: (x['total_bill'].mean() > 18) & (x['tip'].std() < 1.5)
)
print(f"복합 조건 만족 데이터 수: {len(complex_filter)}")
```

#### 2-4. 고급 그룹화 함수 - apply() 함수

```python
# apply() 함수 활용
def analyze_day_performance(group):
    """요일별 성과 분석 함수"""
    result = pd.Series({
        'total_customers': len(group),
        'total_revenue': group['total_bill'].sum(),
        'avg_bill': group['total_bill'].mean(),
        'avg_tip': group['tip'].mean(),
        'tip_percentage': (group['tip'].sum() / group['total_bill'].sum()) * 100,
        'large_party_ratio': (group['size'] >= 4).mean() * 100,
        'peak_hour_sales': group[group['time'] == 'Dinner']['total_bill'].sum()
    })
    return result
```

```python
performance_analysis = df.groupby('day').apply(analyze_day_performance).round(2)
print("요일별 성과 분석:")
display(performance_analysis)
```

```python
# 그룹별 상위 N개 데이터 추출
def top_n_customers(group, n=3):
    """각 그룹에서 상위 N개 고객 추출"""
    return group.nlargest(n, 'total_bill')[['total_bill', 'tip', 'size']]
```

```python
top_customers = df.groupby('day').apply(lambda x: top_n_customers(x, 2))
print("\n요일별 상위 고객:")
display(top_customers)
```

### 5. 데이터 병합 (Merge)
#### 5-1. 기본 병합 예시

```python
# 병합용 샘플 데이터 생성
data1 = {'ID': [1, 2, 3, 4, 5],
         'Name': ['John', 'Emily', 'Charlie', 'Jessica', 'Brian']}
data2 = {'ID': [1, 2, 3, 4, 6],
         'Age': [25, 30, 35, 40, 45]}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

print("데이터프레임 1:")
display(df1)
print("\n데이터프레임 2:")
display(df2)
```

```python
# 다양한 조인 방식
print("\n1. Inner Join:")
inner_result = pd.merge(df1, df2, on='ID', how='inner')
display(inner_result)
```

```python
print("\n2. Left Join:")
left_result = pd.merge(df1, df2, on='ID', how='left')
display(left_result)
```

```python
print("\n3. Right Join:")
right_result = pd.merge(df1, df2, on='ID', how='right')
display(right_result)
```

```python
print("\n4. Outer Join:")
outer_result = pd.merge(df1, df2, on='ID', how='outer')
display(outer_result)
```

#### 5-5. 종합 실습

```python
# 종합 실습: 레스토랑 데이터 분석
print("=== 종합 실습: 레스토랑 데이터 분석 ===")

# 1. 멀티 인덱스로 데이터 구조화
multi_analysis = df.set_index(['day', 'time', 'smoker'])
display(multi_analysis.head())
```

```python
# 2. 그룹별 상세 분석
comprehensive_analysis = df.groupby(['day', 'smoker']).agg({
    'total_bill': ['count', 'mean', 'sum'],
    'tip': ['mean', 'std'],
    'size': ['mean', 'max']
}).round(2)

print("1. 종합 그룹 분석:")
display(comprehensive_analysis)
```

```python
# 3. 피벗 테이블로 요약
summary_pivot = df.pivot_table(
    values=['total_bill', 'tip'],
    index='day',
    columns=['time', 'smoker'],
    aggfunc='mean',
    fill_value=0
).round(2)

print("\n2. 피벗 테이블 요약:")
display(summary_pivot)
```

```python
# 4. 변환 및 정규화
df_normalized = df.copy()
df_normalized['tip_normalized'] = df.groupby('day')['tip'].transform(
    lambda x: (x - x.mean()) / x.std()
)
df_normalized['bill_rank'] = df.groupby('day')['total_bill'].transform('rank')

print("\n3. 정규화 및 순위 결과:")
display(df_normalized[['day', 'tip', 'tip_normalized', 'total_bill', 'bill_rank']].head(10))

print("\n=== 실습 완료 ===")
```
